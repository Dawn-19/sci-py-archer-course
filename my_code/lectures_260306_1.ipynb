## scipy

case I

$$
\int _{x_{min}} ^{x_{max}}\, dx \int _{g(x)} ^{h(x)} f(x,y) \, dy = \int _{-r} ^{r} \int _{-\sqrt(r^2-x^2)} ^{\sqrt(r^2-x^2)} x^2 \, dx\, dy = \pi r^2
$$


```python
import numpy as np
from scipy.integrate import dblquad

def integrand(y,x):
    return 1

def xminlim(x,r):
    return -np.sqrt(r*r - x*x)

def xmaxlim(x,r):
    return np.sqrt(r*r - x*x)

def integrate_to_pi(r):
    result = dblquad(integrand, -1*r, r,
                     lambda x: xminlim(x,r),
                     lambda x: xmaxlim(x,r))
    area, err = result[0], result[1]
    return area/(r*r),err
```

```python
print(integrate_to_pi(1.0))
```

case II:

$$
\int_0^{\pi/2} dx \int_0^1 dy \quad f(x,y)
$$

where $f(x,y) = y sin(x)$. The answer should be 1/2.

```python
import numpy as np
from scipy.integrate import dblquad

def Int_f(y,x):
    return y*np.sin(x)

def yminlim(x):
    return x

def ymaxlim(x):
    return 2*x

def integrate_to_piover2():
    result = dblquad(Int_f, 0, np.pi/2,
                     lambda x: yminlim(x),
                     lambda x: ymaxlim(x))
    area, err = result[0], result[1]
    return area, err
```

```python
print(integrate_to_piover2())
```
